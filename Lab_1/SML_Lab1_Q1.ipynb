```python
artificial_df=pd.read_csv('DP_Artifical.csv')
artificial_df.info()
```

```python
print(artificial_df.isnull().sum())
artificial_df['Age'] = artificial_df['Age'].fillna(artificial_df['Age'].mean())
artificial_df['Income'] = artificial_df['Income'].fillna(-1)
print(artificial_df.info())
```

```python
numeric_cols=[cols for cols in artificial_df.columns if artificial_df[cols].dtype in ['int64','float64']]
# imp = SimpleImputer(missing_values=np.nan, strategy="mean")
# artificial_df[numeric_cols]=imp.fit_transform(artificial_df[numeric_cols])
# print(numeric_cols)
```

```python
artificial_df.dropna(axis=0,inplace=True)
artificial_df.info()
```

```python
for i in numeric_cols:
  print(i,'mean=',artificial_df[i].mean())
```

```python
for i in numeric_cols:
  print(i,'median=',artificial_df[i].median())
```

```python
for i in numeric_cols:
  print(i,'variance=',artificial_df[i].var())
```

```python
for i in numeric_cols:
  print(i,'standard deviation=',artificial_df[i].std())
```

```python
from scipy.stats import norm
for i in numeric_cols:
  # plt.plot(artificial_df[i],norm.pdf(artificial_df[i],artificial_df[i].mean(),artificial_df[i].std()))
  # plt.show()
  print(i,'Normal Distribution=',norm.pdf(artificial_df[i],artificial_df[i].mean(),artificial_df[i].std()))
```
