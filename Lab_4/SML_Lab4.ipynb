```python
import numpy as np
from sympy import *
from sympy.solvers.solveset import solvify,solveset
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# define symbols only
x,y= symbols('x y')

# define the function
y=x/(x**2+1)

# define the lower interval value and upper interval value
lv=[-2,-3,-5,-10,-50,-75,-100,-200,-300,-400,-500,-750,-800,-900,-1000]
# uv=input("Enter upper interval value:")
global_maxy=[]
global_maxx=[]
global_miny=[]
global_minx=[]

# the value of y on the extreme interval value
for j in range(0,len(lv)):
  yleft=y.subs(x,lv[j])
  yright=y.subs(x,-lv[j])

  # print the yleft and y right value
  print("Evaluation if leftvalues:", yleft)
  print("Evaluation if rightvalues", yright)


  MaxMinArray =np.array([[int(lv[j]),yleft],[int(-lv[j]), yright]])
  # to calculate first order derivative and second order derivatives
  expr=y.diff(x)
  expr2=y.diff(x,2)

  # to find the crtical value
  cp=solvify(expr,x,S.Reals)

  # To find the length of the crtical points
  ncp=len(cp)

  print("\n Total number of critical points:", ncp)
  print("The critical points are :", cp)


  for i in range(ncp):
    yv=y.subs(x,cp[i])
    print("\n For critical points: ", cp[i])
    print("The associated y value:", yv)
    print("The point is :(", cp[i],",",yv,")")
    subV=expr2.subs(x,cp[i])

    if(subV > 0):
      print("The", cp[i]," is the local minimum")
      ymin1=y.subs(x,cp[i])
      np_arr2=np.array([[cp[i],ymin1]])
      if(cp[i]== int(lv[j])or cp[i]==int(-lv[j])):
        continue
      MaxMinArray=np.append(MaxMinArray,np_arr2,axis=0)
    elif (subV<0):
      print("The",cp[i],"is the local maxima")
      ymax1= y.subs(x,cp[i])
      np_arr2=np.array([[cp[i],ymax1]])
      if (cp[i]==int(lv[j]) or cp[i]==int(-lv[j])):
        continue
      MaxMinArray= np.append(MaxMinArray, np_arr2,axis=0)
    else:
      print("The",cp[i]," Nothing to say")

      
  forIndex=MaxMinArray[:,1]
  ind=np.where(forIndex==np.amin(forIndex))
  print("\n The global minima on [", lv[j],",",-lv[j],"] is\n", MaxMinArray[ind,1][0][0], "at", MaxMinArray[ind,0][0][0])
  global_maxy+=[MaxMinArray[ind,1][0][0]]
  global_maxx+=[MaxMinArray[ind,0][0][0]]

  ind=np.where(forIndex==np.max(MaxMinArray[:,1]))
  print("\n The global maxima on [", lv[j],",",-lv[j],"] is\n", MaxMinArray[ind,1][0][0], "at", MaxMinArray[ind,0][0][0])
  global_miny+=[MaxMinArray[ind,1][0][0]]
  global_minx+=[MaxMinArray[ind,0][0][0]]
```

```python
plt.scatter(x=global_maxx,y=global_maxy,color='red')
plt.xlabel('x')
plt.ylabel('Global maxima')
```

```python
plt.scatter(x=global_minx,y=global_miny,color='green')
plt.xlabel('x')
plt.ylabel('Global minima')
```
